```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
from scipy.stats.stats import pearsonr
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
df = pd.read_csv('./data/world-happiness_2017.csv')
columns = ['country', 'happiness_rank', 'happiness_score', 'economy_percapita', 'family', 
           'health_life', 'freedom', 'generosity', 'trust_government']
df.columns = columns
df.head()
```

```python
fig, ax = plt.subplots(8, 8, figsize=(25, 25))

for i in range(8):
    for j in range(8):
        x = df[columns[j+1]].values
        y = df[columns[i+1]].values
        p = pearsonr(x, y)[0]        
        ax[i, j].set_title('Correlation %.2f'%p)
        ax[i, j].scatter(x, y, color='#5A88DD', linewidth=.5, edgecolor='k', alpha=.4)
        if j == 0:
            ax[i, j].set_ylabel(columns[i+1])
        if i == 7:
            ax[i, j].set_xlabel(columns[j+1])
plt.tight_layout()
plt.savefig('correlation.png')
```

```python
mat = np.zeros(shape=(8, 8))
plt.figure(figsize=(8, 8))

plt.xticks(np.arange(8), columns[1:], rotation=90)
plt.yticks(np.arange(8), columns[1:])

for i in range(8):
    for j in range(8):
        x = df[columns[j+1]].values
        y = df[columns[i+1]].values
        mat[i][j] = pearsonr(x, y)[0]
for i in range(8):
    for j in range(8):
        color = 'w' if mat[i, j] > 0 else 'k'
        text = plt.text(i, j, '%.2f'%mat[i, j], ha='center', va='center', color=color)
cmap = plt.imshow(mat, cmap='YlGn')
plt.title('Attributes Correlation', fontdict=dict(fontsize=18))
cbar = plt.colorbar(cmap, fraction=0.046, pad=0.04)
cbar.set_label(label='Pearson', rotation=270, )

plt.show()
```
